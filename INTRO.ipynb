```python
import numpy as np
```

```python
##Create a dummy data
Student_info = {
    "Name":["Stephen", "Dora", "Andrew", "Charles", "Jay", "Charlotte", "Mary", "Danny", "Avyana"],
     "Scores":[59, 78, 63, 85, 92, 78, 66, 68, 91]
}
print(Student_info)
```

```python
#Convert into array
name_array=np.array(Student_info['Name'])
score_array=np.array(Student_info['Scores'])
print(name_array)
print(score_array)
```

```python
#Sort the names
sorted_names=np.sort(name_array)
print(sorted_names)
```

```python
#Sort the scorein ascending order
sorted_score = np.sort(score_array)
print(sorted_score)
```

```python
#Sort the score in descending order
sorted_score = np.sort(score_array)[::-1]
print(sorted_score)
```
